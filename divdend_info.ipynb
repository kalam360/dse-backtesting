```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
```

```python
with open("./file.html") as f:
    html = f.read()
    soup = BeautifulSoup(html)
    
```

```python
base = "https://www.investing.com"

links = soup.find_all('a')
dividend_links = []
for each in links:
    url = base + each['href']
    print(url)
    try:
        html = BeautifulSoup(requests.get(url).content).find('a', attrs={'data-test': 'Dividends'})['href']
    except:
        continue
    dividend_links.append(base + html)
    print(base+html)
    time.sleep(5)
```

```python
pd.Series(dividend_links).to_csv("./links.csv")
```
