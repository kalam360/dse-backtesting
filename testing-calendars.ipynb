```python
from exchange_calendars import register_calendar, get_calendar
from dse_bundles.lib.exchange_calendar_xdse import XDSExchangeCalendar
import july
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from glob import glob
```

```python
register_calendar("XDSE", XDSExchangeCalendar())
xdse = get_calendar("XDSE")
```

```python
session = xdse.schedule.loc['2012-01-01':'2021-12-31']
session = [x.date() for x in session.market_close]
files = glob("./before_process_data/csv/*")
for each in files:
    symbol = each.split("/")[-1]
    print(symbol)
    data = pd.read_csv(each, index_col='date', parse_dates=True)
    index = [x.date() for x in data.index]
    to_remove = list(set(index) - set(session))
    print(to_remove)
    data = data[~data.index.isin(to_remove)]
    data.to_csv(f"./dse_bundles/data/daily/{symbol}") 
```
